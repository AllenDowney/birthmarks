```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
from lightkurve import search_lightcurvefile
from requests.exceptions import HTTPError
from astropy.coordinates import SkyCoord
from astroquery.mast import Tesscut
from astropy.timeseries import LombScargle
from photutils import CircularAperture, aperture_photometry

import astropy.units as u
```

```python
coord = SkyCoord('00:19:26.0 +44:01:39', unit=(u.hour, u.deg))
# coord = SkyCoord('12 00 51.0 -46 08 13.0', unit=(u.hour, u.deg))

# slcf = search_lightcurvefile(coord, mission='K2')

# lc = slcf.download_all()
# pdcsap = lc.PDCSAP_FLUX.stitch()
size = 10
cutouts = Tesscut.get_cutouts(coord, size=size)
```

```python
if len(cutouts) > 0:
    times = cutouts[0][1].data['TIME']
    flux = cutouts[0][1].data['FLUX'].sum(axis=(1, 2)) - (size**2) * np.median(cutouts[0][1].data['FLUX'], axis=(1, 2))
    flux /= np.nanmedian(flux)

    mask = (flux > 0.95) & (flux < 1.05)

    ap = CircularAperture([6, 6], 3)
#     ap = CircularAperture([6, 6], 2)
    fluxes = []

    for j, image in enumerate(cutouts[0][1].data['FLUX']):
        if mask[j]:
            f = aperture_photometry(image - np.median(image), ap)

            fluxes.append(f['aperture_sum'].data[0])

    fluxes = np.array(fluxes)

    dip_mask = fluxes > np.percentile(fluxes, 5)

    fluxes = np.interp(times[mask], times[mask][dip_mask], fluxes[dip_mask])

    detrended = fluxes / np.polyval(np.polyfit(times[mask] - times[mask].mean(), fluxes, 2), times[mask] - times[mask].mean())
```

```python
time = times[mask] #pdcsap.time
flux = detrended #pdcsap.flux

notnanflux = ~np.isnan(flux)
flux = flux[notnanflux]# & (time > 1326) & (time < 1337)]
time = time[notnanflux]# & (time > 1326) & (time < 1337)]

# flux = flux[notnanflux & (time > 1341.5) & (time < 1347)]
# time = time[notnanflux & (time > 1341.5) & (time < 1347)]
flux /= np.mean(flux)
# pdcsap.plot()
```

```python
periods = np.linspace(1, 10, 1000) * u.day
freqs = 1 / periods
powers = LombScargle(time[~np.isnan(flux)] * u.day, flux[~np.isnan(flux)]).power(freqs)
best_period = periods[powers.argmax()]
```

```python
from fleck import Star
from emcee import EnsembleSampler
from multiprocessing import Pool
from corner import corner

u_ld = [0.46, 0.11]
contrast = 0.7
phases = (time % best_period.value) / best_period.value
s = Star(contrast, u_ld, n_phases=len(time), rotation_period=best_period.value)
```

```python
init_lons = np.array([0, 150, 250])
init_lats = [0, 0, 0]
init_rads = [0.01, 0.2, 0.2]

yerr = 0.005

init_p = np.concatenate([init_lons, init_lats, init_rads])
```

```python
p = init_p.copy()

lons = p[0:3]
lats = p[3:6]
rads = p[6:9]

lc = s.light_curve(lons[:, None] * u.deg, lats[:, None] * u.deg, rads[:, None], 
                   inc_stellar=90*u.deg, times=time, time_ref=0)

lc /= np.mean(lc)

plt.plot(time, flux)
plt.plot(time, lc)
```

```python
def log_likelihood(p):
    lons = p[0:3]
    lats = p[3:6]
    rads = p[6:9]
    
    lc = s.light_curve(lons[:, None] * u.deg, lats[:, None] * u.deg, rads[:, None], 
                       inc_stellar=90*u.deg, times=time, time_ref=0)[:, 0]
    
#     plt.plot(time, lc/lc.max())
#     plt.plot(time, flux)
#     plt.show()
    return - 0.5 * np.sum((lc/np.mean(lc) - flux)**2 / yerr**2)
    
def log_prior(p): 
    lons = p[0:3]
    lats = p[3:6]
    rads = p[6:9]
    if (np.all(rads < 0.8) and np.all(rads > 0) and np.all(lats > -60) and 
        np.all(lats < 60) and np.all(lons > 0) and np.all(lons < 360)): 
        return 0
    return -np.inf

def log_probability(p):
    lp = log_prior(p)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(p)

ndim = len(init_p)
nwalkers = 2 * ndim
nsteps = 50000

pos = []

while len(pos) < nwalkers: 
    trial = init_p + 0.01 * np.random.randn(ndim)
    lp = log_prior(trial)
    if np.isfinite(lp):
        pos.append(trial)
```

```python
plt.plot(time, flux)

for p in pos: 
    
    lons = p[0:3]
    lats = p[3:6]
    rads = p[6:9]
    lc = s.light_curve(lons[:, None] * u.deg, lats[:, None] * u.deg, rads[:, None], 
                       inc_stellar=90*u.deg, times=time)[:, 0]

    lc /= np.mean(lc)

    plt.plot(time, lc, color='k')
```

```python
with Pool() as pool:
    sampler = EnsembleSampler(nwalkers, ndim, log_probability, pool=pool)
    sampler.run_mcmc(pos, nsteps, progress=True);

# sampler = EnsembleSampler(nwalkers, ndim, log_probability)
# sampler.run_mcmc(pos, nsteps, progress=True);
```

```python
samples_burned_in = sampler.flatchain[len(sampler.flatchain)//2:, :]

np.save('data/qatar4_samples.npy', samples_burned_in)

corner(samples_burned_in);
```

```python
import healpy as hp

fig, ax = plt.subplots(1, 3, figsize=(8, 1))    
ax[0].plot(time, flux, ',', color='k')

for i in np.random.randint(0, len(samples_burned_in), size=50): 
    
    trial = samples_burned_in[i, :]
        
    lons = trial[0:3]
    lats = trial[3:6]
    rads = trial[6:9]
    
    lc = s.light_curve(lons[:, None] * u.deg, lats[:, None] * u.deg, rads[:, None], 
                       inc_stellar=90*u.deg, times=time, time_ref=0)[:, 0]
    ax[0].plot(time, lc/lc.mean(), color='DodgerBlue', alpha=0.05)

f_S = np.sum(samples_burned_in[:, -3:]**2 / 4, axis=1)
    
ax[1].hist(f_S, bins=25, histtype='step', lw=2, color='k', range=[0, 0.12], density=True)
ax[0].set(xlabel='BJD - 2454833', ylabel='Flux', ylim=[0.98, 1.02])
ax[1].set_xlabel('$f_S$')

NSIDE = 2**10

NPIX = hp.nside2npix(NSIDE)

m = np.zeros(NPIX)

for lon, lat, rad in np.median(samples_burned_in, axis=0).reshape((3, 3)).T:
    t = np.radians(lat + 90)
    p = np.radians(lon)
    spot_vec = hp.ang2vec(t, p)
    ipix_spots = hp.query_disc(nside=NSIDE, vec=spot_vec, radius=rad)
    m[ipix_spots] = 0.7

cmap = plt.cm.Greys
cmap.set_under('w')

plt.axes(ax[2])
hp.mollview(m, cbar=False, title="", cmap=cmap, hold=True,
            max=1.0, notext=True, flip='geo')
hp.graticule(color='silver')

fig.suptitle('Qatar-4')

for axis in ax: 
    for sp in ['right', 'top']:
        axis.spines[sp].set_visible(False)
fig.savefig('plots/Qatar4.pdf', bbox_inches='tight')
```
