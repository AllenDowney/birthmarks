```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
from lightkurve import search_lightcurvefile
from requests.exceptions import HTTPError
from astropy.coordinates import SkyCoord
from astroquery.mast import Tesscut
from astropy.timeseries import LombScargle
from photutils import CircularAperture, aperture_photometry

import astropy.units as u
```

```python
table = Table.read('../data/gagne2018.vot')
```

```python
from astroquery.mast import Conf
from scipy.ndimage import gaussian_filter1d

Conf.timeout = 1

size = 10

smoothed_amps_usco = []

usco = table[np.array([i.startswith('F') or i.startswith('G') or i.startswith('K')
                      for i in table['SpT']]) & (table['Assoc'] == b'USCO')] 
usco_periods = []
```

```python
coords = SkyCoord(usco['RAJ2000'], usco['DEJ2000'], unit=(u.hour, u.deg))
binary_list = [3, 5, 20] # [1, 24]
limit_pmax_list = [] # [5]
save_list = [25] #[41]

usco_smamps = []

usco_periods = []

for i, coord in enumerate(coords):

    slcf = search_lightcurvefile(coord, mission='K2')
    lc = slcf.download_all()

    if lc is not None and i not in binary_list:
        print('coord', i)
        if len(lc) > 1: 
            lc = lc[0]
        pdcsap = lc.PDCSAP_FLUX if not hasattr(lc.PDCSAP_FLUX, 'stitch') else lc.PDCSAP_FLUX.stitch()
        mask = ~np.isnan(pdcsap.flux) & (pdcsap.flux / np.nanmedian(pdcsap.flux) > 0.8) & (pdcsap.flux / np.nanmedian(pdcsap.flux) < 1.2)
        times = pdcsap.time[mask] 
        detrended = pdcsap.flux[mask] / np.nanmedian(pdcsap.flux)
        detrended = detrended / np.polyval(np.polyfit(times - times.mean(), detrended, 2), times - times.mean())
        
        # L-S periodogram
        ls = LombScargle(times * u.day, detrended)
        periods = np.linspace(0.05, 
                              15 if i not in limit_pmax_list else 1, 
                              1000) * u.day
        freq = 1/periods
        power = ls.power(freq)
        best_period = periods[power.argmax()].to(u.day).value

        phases = (((times - 0) % best_period) / best_period)

        smoothed = gaussian_filter1d(detrended[np.argsort(phases)], 50, mode='wrap')

        usco_periods.append(best_period)
        usco_smamps.append(smoothed.ptp())
        
        fig, ax = plt.subplots(1, 4, figsize=(8, 2))
        ax[0].axis('off')
        ax[1].plot(times, detrended, 'k')
        ax[1].set(title='Light curve', xlabel='Time [d]', ylabel='Flux')

        ax[2].plot(periods, power, 'k')
        ax[2].set(title='L-S Periodogram', xlabel='Period [d]', ylabel='Power')

        ax[3].plot(phases, detrended, ',k')
        ax[3].plot(phases[np.argsort(phases)], smoothed, 'r')
        ax[3].set(title='Phase folded LC', xlabel='Phase', ylabel='Flux')

        for axis in ax[1:]:
            for s in ['right', 'top']: 
                axis.spines[s].set_visible(False)

        plt.tight_layout(w_pad=0.2)

        if i in save_list: 
            plt.savefig('plots/usco.pdf', bbox_inches='tight')

        plt.show()
        
#         phases = (((pdcsap.time - 0) % prot) / prot)

#         not_nan = ~np.isnan(pdcsap.flux)
#         smoothed = gaussian_filter1d(pdcsap.flux[not_nan][np.argsort(phases[not_nan])], 250, mode='wrap')

#         plt.plot(phases, pdcsap.flux, ',', alpha=0.5)
#         plt.plot(phases[not_nan][np.argsort(phases[not_nan])], smoothed, '.r')
#         plt.ylim(np.nanpercentile(pdcsap.flux, [1, 99]))
#         plt.show()
#             smoothed_amps_usco.append(smoothed.ptp())
```
